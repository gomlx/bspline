```go
import (
    "github.com/gomlx/bsplines"
    bsplinesplot "github.com/gomlx/bsplines/plotly"
)

%%
b := bsplines.New(2, []float64{0, 1, 2, 3}).
    WithControlPoints([]float64{0, 0, 1, 0, 0})  // len(controls) = len(knots)+degree-1
plot := bsplinesplot.New(b)
plot.Plot()
```

## Randomly Generated Curve

```go
import "math/rand/v2"

%%
b := bsplines.NewRegular(2, 8)
controlPoints := make([]float64, b.NumControlPoints())
controlPoints[0] = 1.0
for ii := range controlPoints {
    if ii > 0 {
        controlPoints[ii] = controlPoints[ii-1] + (2.0*rand.Float64()-1.0)
    }
}
b = b.WithControlPoints(controlPoints)
bsplinesplot.New(b).Plot()
```
